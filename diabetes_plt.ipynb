```python
import pandas as pd
import numpy as np

path = '/content/diabetes.csv'
```

```python
# DataFrame으로 읽기
df = pd.read_csv(path)

df
```

```python
df2=df.drop('Outcome',axis=1)
df.head()
```

```python
y=df2['BMI']
y.value_counts()
y

X=df2.drop('BMI',axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))
```

```python
def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))
```

```python
def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.scatter(y_test, ly_preds, c='y', label='Predicted BMI vs Actual BMI')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect Prediction')
plt.xlabel('Actual BMI (y_test)')
plt.ylabel('Predicted BMI (ly_preds)')
plt.title('Actual vs Predicted BMI values')
plt.legend()
plt.grid(True)
plt.show()
```
